### Introduction to Basic Crosstabs

[Back to Contents](#Contents)

Crosstabs are a mainstay in survey data analysis.  They tend to be overworkd and overused, if not abused.  Too many analysts rely exclusively on tabs.  A client once told me:

> *Everything you need to know is in the tabs*.

Creating tabs is easy with the Pandas *crosstab* function.

```python
##
## Crosstab of Age and Gender
##
xtab = pd.crosstab( df.Age, df.Gender )
xtab.style.set_caption( 'Age-Gender Crosstab' ).\
           bar( color = 'lightgreen',align = 'zero' )
```

**Code Explanation**

Frequency counts are not useful because they depend on the current sample size.  Comparisons across surveys, as in a tracking study, are not possible because of changing sample sizes.  Percents, as normalized frequencies, are better because of a common base.  The *normalize = True* argument does the normalization.  The default is *normalize = False*.

```python
##
## Crosstab of Age and Gender, normalized
##
xtab = pd.crosstab( df.Age, df.Gender, normalize = True )
xtab.style.set_caption( 'Age-Gender Normalized Crosstab' ).\
    format( format ).\
    bar( color = 'lightgreen',align = 'zero' )
```

### Plot a Crosstab

[Back to Contents](#Contents)

The current crosstab data are in *wide-form*: many columns (the width), few rows.  Many analytic procedures require *long-form*: many rows, few columns.  We need to convert the crosstab table to long-form.  This is done by *stacking* the rows.

```python
##
## Stack the data: move from wide-form to long-form
##
df_stacked = xtab.stack().reset_index().rename( columns = { 0:'Percent' } )
df_stacked.head().style.set_caption( 'Stacked Age-Gender Normalized Crosstab' ).\
format( { 'Percent':format }  )
```

**Code Explanation**

The *stack* method stacks each row of the crosstab, one on top of the next.  The column labels of the original crosstab become entries in a new column and the row values because a new value column.

```python
##
## Plot the stacked crosstab with grouped bar chart
##
base = 'Base: All consumers; n = ' + str( df.shape[ 0 ] ) + \
'\nQuestions: What is your gender? What is your age?'
##
ax = sns.barplot( y = 'Age', x = 'Percent', hue = 'Gender', data = df_stacked )
ax.set_title( 'Gender by Age Group', fontsize = font_title )
ax.set( xlabel = 'Percent' )
tick_labels( 'x' )
ax.legend( loc = 1, fontsize = 'small', ncol = 2 )
footer;
```

```python
##
## A heatmap may be better
##
base = 'Base: All consumers; n = ' + str( df.shape[ 0 ] ) + \
'\nQuestions: What is your gender? What is your age?'
##
ax = sns.heatmap( xtab, cmap = "YlGnBu" )
ax.set_title( 'Gender by Age Group\nHeatmap', fontsize = font_title )
ax.annotate( base, (0,0), (0, -0.3), xycoords = 'axes fraction' );
```

#### CATA Analysis

[Back to Contents](#Contents)

*Check All That Apply* (*CATA*) questions are very common in survey work.  The yogurt study had one about the flavors purchased.  We already saw one way to look at the flavors purchased: just count the total.  But this provided no information about individual flavors.

```python
##
## Specify flavor list
##
lst = [ 'Banana', 'Blueberry', 'Boysenberry', 'Strawberry', 'Vanilla', 
       'Coffee', 'Raspberry' ]
n = len( lst )
flavors = df[ lst ].copy()
##
## Recode CATA responses which are currently Yes/No to 1/0
##
for i in range( n ):
    flavors[ lst[ i ] ] = flavors[ lst[ i ] ].apply( lambda x: 1 if x == 'Yes' else 0 )
##
## Calculate some needed quantities: 
##
x = flavors.sum( axis = 0 )  ## sum for each respondent
tot_responses = sum( x )     ## total all repsondents
n_size = df.shape[ 0 ]       ## sample size
##
## Store results in DataFrame
##
df_results = pd.DataFrame()
df_results[ 'Freq' ] = x
df_results[ 'Share' ] = x/tot_responses
df_results[ 'Rate' ] = x/n_size
## 
## Display the results
##
format_dict = { 'Share': format, 'Rate': format }
df_results.style.format( format_dict ).\
    set_caption( 'Flavor Summary: Total Responses = ' + str( tot_responses ) +\
                '; Total Cases: ' + str( n_size )  ).\
    bar( color = 'lightgreen', align = 'zero' )
```

**Interpretation**

The *Rate* is the percent of cases while the *Share* is the percent of responses or total ticks; focus on *Rate* since *Share* does not make sense.

| Term            | Definition                        |
|-----------------|-----------------------------------|
| Cases           | People or Respondents             |
| Total Responses | Sum of Freq                       |
| Freq            | Number of Flavors Selected        |
| Share           | $\frac{Freq}{Total~Response}$     |
| Rate            | $\frac{Freq}{Total~Cases}$        |


### Core Question 2: Brand - Flavor Relationship

[Back to Contents](#Contents)

Frequency counts and statistical tests can be used.

```python
##
## Specify flavor list
##
lst = [ 'Brand', 'Banana', 'Blueberry', 'Boysenberry', 'Strawberry', 'Vanilla', 
       'Coffee', 'Raspberry' ]
flavors = df[ lst ].copy()
##
## Recode CATA responses which are currently Yes/No to 1/0
##
for i in range( 1, 8 ):
    flavors[ lst[ i ] ] = flavors[ lst[ i ] ].apply( lambda x: 1 if x == 'Yes' else 0 )
##
## Pivot table and store results
##
df_results = flavors.pivot_table( index = 'Brand', aggfunc = 'sum', margins = False )
df_results[ 'Responses' ] = df_results.sum( axis = 1 )
df_results[ 'Cases' ] = flavors.Brand.value_counts()
df_results.style.set_caption( 'Frequencies of Cases by Brand' ).\
bar( color = 'lightgreen', align = 'zero' )
```

**Code Explanation**

The *pivot_table* method is like the *crosstab* method but it allows an aggregation specification.  The aggregation is *sum* in this case.

```python
##
## Repeat as Percents
##
lst = [ 'Brand', 'Banana', 'Blueberry', 'Boysenberry', 'Strawberry', 'Vanilla',
       'Coffee', 'Raspberry' ]
n = len( lst )
format_dict = { n:format for n in lst }
flavors = df[ lst ].copy()
##
## Recode CATA responses which are currently Yes/No to 1/0
##
for i in range( 1, n ):
    flavors[ lst[ i ] ] = flavors[ lst[ i ] ].apply( lambda x: 1 if x == 'Yes' else 0 )
##
## Store results
##
df_results = flavors.pivot_table( index = 'Brand', aggfunc = 'mean' )
df_results.style.set_caption( 'Percent of Cases by Brand' ).\
format( format_dict ).bar( color = 'lightgreen', align = 'zero' )
```

**Code Explanation**

The mean is used for aggregation since since the mean of 0/1 values is a proportion.

Do a Pearson Chi-sqaure Test of Independence.  See [here](https://en.wikipedia.org/wiki/Pearson%27s_chi-squared_test) for background.

```python
##
## Pearson Chi-square Test of Independence
##
lst = [ 'Brand', 'Banana', 'Blueberry', 'Boysenberry', 'Strawberry', 'Vanilla', 'Coffee', 'Raspberry' ]
n = len( lst )
format_dict = { 'P-Value': p_value}
##
tmp = df[ lst ].copy()
##
## Define empty lists for chi-square results
##
lst_chi2 = []
lst_pval = []
lst_sig = []
##
## Pearson Chi-square tests of independence
##
for i in range( 1, n ):
    freq = pd.crosstab( tmp.Brand, tmp[ lst[ i ] ] )
    chi2, p, dof, ex = st.chi2_contingency( freq )
    lst_chi2.append( chi2 )
    lst_pval.append( p )
    if p <= 0.05:
        sig = 'Significant'
    else:
        sig = 'Insignificant'
    lst_sig.append( sig )
lst.pop( 0 )
data = { 'Flavor':lst, 'ChiSquare':lst_chi2, 'P-Value':lst_pval, 'Significance':lst_sig }
df_chisq = pd.DataFrame( data )
df_chisq.style.set_caption( 'Pearson Chi Square Homogeneity Test of Each Flavor' )\
.bar( 'ChiSquare', color = 'lightgreen', align = 'zero' ).format( format_dict )\
.applymap( lambda x: 'background-color : yellow' if x == 'Significant' else '' )
```

Cochrane's Q Test is used when the variables have 0/1 values.  See [here](https://en.wikipedia.org/wiki/Cochran%27s_Q_test) for background.

```python
##
## Use Cochrane's Q 
##
##
## Subset flavors
##
lst = [ 'Banana', 'Blueberry', 'Boysenberry', 'Strawberry', 'Vanilla', 
       'Coffee', 'Raspberry' ]
n = len( lst )
flavors = df[ lst ].copy()
##
## Do test
##
( Q, pVal ) = cochrans_q( flavors )
print('\nCOCHRAN\'S Q\n' + '-'*11 )
print('Q = {0:5.3f}, p = {1:5.3f}'.format( Q, pVal ) )
if pVal < 0.05:
    print("\tThere is a significant difference between the flavors.")
else:
    print("\tThere is no significant difference between the flavors.")
```

### Core Question 3: Brand - Segment Relationship

### Core Question 4: Price Elasticity Analysis

[Back to Contents](#Contents)

What is the elasticity of demand for units of yogurt purchased with respect to the average price paid?

```python
##
## Subset the data
##
lst = [ 'Units', 'Price', 'Age', 'Gender' ]
df_model = df[ lst ].copy()
df_model.head().style.set_caption( 'Modeling DataFrame' )
```

```python
##
## Model units purchased as function of price, gender, age
##
## ===> Step 1: Specify model <===
##
formula = 'np.log(Units) ~ np.log(Price) + C(Gender) + C(Age)'
##
## ===> Step 2: Instantiate model <===
##
mod = smf.ols( formula, data = df_model )
##
## ===> Step 3: Fit model <===
##
reg01 = mod.fit()
##
## ===> Step 4: Summarize fit <===
##
print( reg01.summary() )
```

```python
##
## Rerun with Age removed
##
formula = 'np.log(Units) ~ np.log(Price) + C(Gender) + np.log(Price)*C(Gender)'
mod = smf.ols( formula, data = df_model )
reg02 = mod.fit()
print( reg02.summary() )
```

```python
##
## Specify hypothesis for Age
##
hypothesis = 'C(Age)[T.36 to 40 years old] = 0, C(Age)[T.41 to 50 years old] = 0, \
C(Age)[T.51 to 60 years old] = 0, C(Age)[T.61 to 67 years old] = 0, \
C(Age)[T.68 years old and over]'
##
## Run F-Test
##
f_test = reg01.f_test( hypothesis )
##
## Retrieve the p-value
##
pval = round( float( f_test.pvalue ), 4 )
##
## Print results
##
print( '\np-value for F-Test: {}'.format( pval ) )
if pval < 0.05:
    print( '\nSignificant so reject H0' )
else:
    print( '\nInsignificant so do not reject H0' )
```

### Calculate Price Elasticities

```python
##
## Price Elasticities
##
eta_male = reg02.params[ 2:4 ].sum()
eta_female = reg02.params[ 2 ]
data = { 'Gender':[ 'Female', 'Male' ], 'Elasticity':[ eta_female, eta_male ] }
pd.DataFrame( data ).style.set_caption( 'Elasticity Summary' ).\
bar( color = 'lightgreen', align = 'mid' )
```

**Interpretation**

The price elasticity is the coefficient for the logged price variable: -1.4, but this has to be adjusted for males and females.  The adjustment is given by the Gender dummy variable, *np.log(Price):C(Gender)[T.Male]*.  This dummy is 1 for males and 0 of females.  For males, the elasticity is $-1.4 + (- 0.09) = -1.5$ and for females it is just -1.4.  For males, if the price rises 1\%, unit sales fall 1.5%.  This indicates that yogurts are highly elastic for males.  On average, males are more price elastic than females.  This should be expected since women are more health conscious.  Revenue will also change.  If price rises, revenue from both genders falls.  The amount revenue falls (in percentage terms) is given by $1 + elasticity$.  So for a 1% rise in price, revenue from males will fall 0.5% ($= 1 + [-1.5]$).
